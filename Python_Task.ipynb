# Task 4: NumPy Array Manipulation

```python
import numpy as np

# Create two sample NumPy arrays
array1 = np.array([[1, 2, 3],
                   [4, 5, 6]])

array2 = np.array([[7, 8, 9],
                   [10, 11, 12]])

concatenated_vertically = np.concatenate((array1, array2), axis=0)

concatenated_horizontally = np.concatenate((array1, array2), axis=1)
print("Array 1:")
print(array1)
print("\nArray 2:")
print(array2)
print("\nConcatenated Vertically:")
print(concatenated_vertically)
print("\nConcatenated Horizontally:")
print(concatenated_horizontally)
```

# Task 5: Pandas Data Filtering

```python
import pandas as pd

# Sample sales dataset
data = {
    'Date': ['2023-09-01', '2023-09-02', '2023-09-03', '2023-09-04', '2023-09-05'],
    'Product': ['A', 'B', 'A', 'C', 'B'],
    'Sales': [100, 150, 120, 80, 200]
}

df = pd.DataFrame(data)
df['Date'] = pd.to_datetime(df['Date'])

start_date = '2023-09-02'
end_date = '2023-09-04'
filtered_df = df.loc[(df['Date'] >= start_date) & (df['Date'] <= end_date)]

print(filtered_df)
```

# Task 7: NumPy Statistical Analysis

```python
import numpy as np

# Sample dataset with two variables, X and Y
X = np.array([1, 2, 3, 4, 5])
Y = np.array([2, 3, 4, 5, 6])

# Calculate variance and standard deviation for X
variance_X = np.var(X)
std_deviation_X = np.std(X)

# Calculate variance and standard deviation for Y
variance_Y = np.var(Y)
std_deviation_Y = np.std(Y)

# Calculate the correlation coefficient between X and Y
correlation_coefficient = np.corrcoef(X, Y)[0, 1]

# Display the results
print("Variance of X:", variance_X)
print("Standard Deviation of X:", std_deviation_X)

print("\nVariance of Y:", variance_Y)
print("Standard Deviation of Y:", std_deviation_Y)

print("\nCorrelation Coefficient between X and Y:", correlation_coefficient)
```

# Task 8: Pandas Data Grouping

```python
import pandas as pd

data = {
    'Date': ['2023-09-01', '2023-09-02', '2023-09-03', '2023-09-04', '2023-09-05'],
    'Region': ['North', 'South', 'North', 'East', 'West'],
    'SalesAmount': [1000, 1500, 1200, 800, 2000]
}

df = pd.DataFrame(data)
grouped_data = df.groupby('Region')['SalesAmount'].agg(['mean', 'median'])
print(grouped_data)
```

# Task 10: Pandas Data Visualization

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May'],
    'Rent': [1500, 1500, 1500, 1550, 1550],
    'Utilities': [200, 220, 210, 220, 225],
    'Groceries': [400, 410, 390, 405, 410]
}

df = pd.DataFrame(data)
plt.figure(figsize=(10, 4))
plt.bar(df['Month'], df['Rent'], label='Rent', color='skyblue')
plt.bar(df['Month'], df['Utilities'], label='Utilities', color='lightgreen', bottom=df['Rent'])
plt.bar(df['Month'], df['Groceries'], label='Groceries', color='lightcoral', bottom=df['Rent'] + df['Utilities'])
plt.xlabel('Month')
plt.ylabel('Expense Amount ($)')
plt.title('Monthly Household Expenses')
plt.legend()

plt.figure(figsize=(6, 6))
explode = (0.1, 0.1, 0.1)
plt.pie(df.loc[df['Month'] == 'May'][['Rent', 'Utilities', 'Groceries']].values[0], labels=['Rent', 'Utilities', 'Groceries'], explode=explode, autopct='%1.1f%%', shadow=True, colors=['skyblue', 'lightgreen', 'lightcoral'])
plt.title('Expense Categories in May')

plt.figure(figsize=(6, 4))
plt.scatter(df['Rent'], df['Utilities'], c='blue', label='Rent vs. Utilities', marker='o')
plt.xlabel('Rent ($)')
plt.ylabel('Utilities ($)')
plt.title('Correlation between Rent and Utilities')

plt.tight_layout()
plt.show()
```
